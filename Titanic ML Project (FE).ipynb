## Titanic ML Project

### The Challenge

The sinking of the Titanic is one of the most infamous shipwrecks in history.

On April 15, 1912, during her maiden voyage, the widely considered “unsinkable” RMS Titanic sank after colliding with an iceberg. Unfortunately, there weren’t enough lifeboats for everyone onboard, resulting in the death of 1502 out of 2224 passengers and crew.

While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others.

In this challenge, we ask you to build a predictive model that answers the question: “what sorts of people were more likely to survive?” using passenger data (ie name, age, gender, socio-economic class, etc).

### Variance Inflation Factor

```python
# calculate Variance Inflation Factor for each explanatory variable
from statsmodels.stats.outliers_influence import variance_inflation_factor

def VIF(df, columns):
    values = sm.add_constant(df[columns]).values
    num_columns = len(columns)+1
    vif = [variance_inflation_factor(values, i) for i in range(num_columns)]
    return pd.Series(vif[1:], index=columns)
```

### Ridge Regression

We want to isolate our top features.

```python
def one_standard_error_rule(model, results, param_grid, n_splits, neg_mean_squared_error=True):
    
    assert neg_mean_squared_error == True # function is defined specifically for neg_mean_squared_error
    
    range_x = param_grid # results['param_'+list(param_grid.keys())[0]].data
    std_vs_x  = pd.Series(results['std_test_score'], index = range_x)
    sem_vs_x  = std_vs_x/np.sqrt(n_splits)
    
    mean_vs_x = pd.Series(results['mean_test_score'], index = range_x)        
    mean_vs_x = mean_vs_x*(-1)
    
    x_min = mean_vs_x.idxmin()
    sem = sem_vs_x[x_min]
    
    if (model=='pcr'):
        x_1se = mean_vs_x[mean_vs_x <= min(mean_vs_x) + sem].index.min()
    elif (model=='ridge') | (model=='lasso'):
        x_1se = mean_vs_x[mean_vs_x <= min(mean_vs_x) + sem].index.max()
        
    #x_1se_idx = int(np.argwhere(range_x == x_1se)[0])
    
    return x_min, x_1se
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
```

```python
alpha_max = 10**5
rr = Ridge(alpha=alpha_max, random_state=88)
rr.fit(train_enc_X, train_enc_Y)
```

```python
plt.plot(rr.coef_)
plt.show()
print(max(abs(rr.coef_)))
```

```python
alpha_grid = np.logspace(-1, 5, num=50, base=10)
coefs = []
```

```python
for a in alpha_grid:
    rr = Ridge(alpha=a, fit_intercept=False, random_state=88)
    rr.fit(train_enc_X, train_enc_Y)
    coefs.append(rr.coef_)

plt.figure(figsize=(12, 10))
ax = plt.gca()
ax.plot(alpha_grid, coefs)
ax.set_xscale('log')
plt.xlabel('Alpha')
plt.ylabel('Coefficients')
plt.title('Ridge coefficients as a function of the regularization')
plt.show()
```

### Ridge Hyper-parameter Tuning

```python
alpha_grid = {'alpha': np.logspace(-1, 5, num=50, base=10)}

rr = Ridge(random_state=88)
rr_cv = GridSearchCV(rr, alpha_grid, scoring='neg_mean_squared_error', cv=10)
rr_cv.fit(train_enc_X, train_enc_Y)
```

```python
range_alpha = rr_cv.cv_results_['param_alpha'].data
MSE_scores = rr_cv.cv_results_['mean_test_score']*(-1)
x_min, x_1se = one_standard_error_rule(model='ridge',
                                       results=rr_cv.cv_results_,
                                       param_grid=range_alpha,
                                       n_splits=10,
                                       neg_mean_squared_error=True)
plt.figure(figsize=(8, 6))
ax = plt.gca()
ax.set_xscale('log')
plt.xlabel('Alpha', fontsize=16)
plt.ylabel('CV MSE', fontsize=16)
plt.scatter(range_alpha, MSE_scores, s=30)
plt.axvline(x=x_min, color='m')
plt.axvline(x=x_1se, color='c')
plt.grid(True, which='both')

plt.tight_layout()
plt.show()
```

```python
print('Alpha one standard error rule:', x_1se)
```

#### Most Frequent Name Titles By Class

```python
pclass1 = train[train['Pclass'] == 1]
pclass2 = train[train['Pclass'] == 2]
pclass3 = train[train['Pclass'] == 3]
```

```python
#Most frequent name titles words by class
from collections import Counter

def most_frequent(data, number, pclass):
    pclass1 = data[data['Pclass'] == 1]
    pclass2 = data[data['Pclass'] == 2]
    pclass3 = data[data['Pclass'] == 3]
    
    pclass1_most_freq = Counter(" ".join(pclass1["Name"]).split()).most_common(number)
    pclass2_most_freq = Counter(" ".join(pclass2["Name"]).split()).most_common(number)
    pclass3_most_freq = Counter(" ".join(pclass3["Name"]).split()).most_common(number)
    
    if pclass == 1:
        return pclass1_most_freq
    elif pclass == 2:
        return pclass2_most_freq
    else:
        return pclass3_most_freq
```

```python
one_top_names = most_frequent(train, 300, 1)
one_top_names
```

```python
two_top_names = most_frequent(train, 300, 2)
two_top_names
```

```python
three_top_names = most_frequent(train, 300, 3)
three_top_names
```

```python
# Extracting just the words and names from the tuple
[item[0] for item in one_top_names]
```

```python
#  Most Common Words only in a certain class and not the others.
def only_class(pclass, one_top, two_top, three_top):
    
    pclass1_names = [item[0] for item in one_top]
    pclass2_names = [item[0] for item in two_top]
    pclass3_names = [item[0] for item in three_top]
    
    if pclass == 1:
        pclass1_only = []
        for i in range(len([pclass1_names])):
            if pclass1_names[i][0] not in pclass2_names and pclass1_names[i][0] not in pclass3_names:
                pclass1_only.append(pclass1_names[i][0])
        return pclass1_only
    elif pclass == 2:
        pclass2_only = []
        for i in range(len([pclass2_names])):
            if pclass2_names[i][0] not in pclass1_names and pclass2_names[i][0] not in pclass3_names:
                pclass2_only.append(pclass2_names[i][0])
        return pclass2_only
    else:
        pclass3_only = []
        for i in range(len([pclass3_names])):
            if pclass3_names[i][0] not in pclass1_names and pclass3_names[i][0] not in pclass2_names:
                pclass3_only.append(pclass3_names[i][0])
        return pclass3_only
```

```python
only_class(2, one_top_names, two_top_names, three_top_names)
```

```python
only_class(3, one_top_names, two_top_names, three_top_names)
```
